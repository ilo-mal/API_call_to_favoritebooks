<h2>Favoritebooks web Api call to display ten books with higest average rating:</h2>

```python
import requests
import pandas as pd
response = requests.get('https://favoritebooksweb.herokuapp.com/main/api/')
book_data = pd.DataFrame(response.json())
print(book_data.head(10))
```

<h3>Chart by book genre</h3>

```python
import matplotlib.pyplot as plt
genre = book_data['title'].groupby(book_data['genre']).count()
fig = plt.figure(figsize=(20, 8))
plt.pie(genre, labels=genre.index, autopct='%.2f%%')
plt.show()
```
